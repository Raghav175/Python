### Question 2: Write a Python program to reverse a string without using [::-1]

- `sample input : `

```python
# Please write The solution Here for above question 2:

txt = "Sehwag"
rev_s = ""
l = len(txt)

for i in range(l):
  rev_s = rev_s + txt[(l-1)-i]

print(rev_s)
```

### Question 3: Write a Python program to check if a string is a palindrome don't use [::-1].

- `sample input : `
    - `var = 'NUN'`
- `sample output :`
    - `result : It is a Palindrome`
    
- `palindrome concept = https://www.dictionary.com/e/palindromic-word/#:~:text=A%20palindrome%20is%20a%20word,roots%20in%20the%20early%201600s.`

```python
# Please write The solution Here for above question 3:

txt = 'radar'
rev_s = ""
l = len(txt)

for i in range(l):
  rev_s = rev_s + txt[(l-1)-i]

if (txt == rev_s):
  print(f"{txt} is a palindrome")
else:
  print(f"{txt} is not a palindrome")
```

### Question 4: Write a Python program to find the most common character in a string.

- `sample input : `
    - `var = 'Hello world '`
- `sample output :`
    - `result : l, because l is repeating 3 times `

```python
# Please write The solution Here for above question 4:

txt = 'Hello worold'
cc = {}

for char in txt:
  if char not in cc:
    cc[char] = 1
  else:
    cc[char] = cc[char] + 1

#print (cc)
#print(max(cc.values()))
max_key = max(cc, key=cc.get)
print(max_key)
```

### Question 5: Write a Python program to check if two strings are anagrams.

- `sample input : `
    - `var1 = 'listen'`
    - `var2 = 'silent'`
- `sample output :`
    - `result : both are anagram because characters in var2 should present in var1 `

```python
# Please write The solution Here for above question 5:

txt1 = 'earth'
txt2 = 'heart'

txt_s1 = set(txt1)
txt_s2 = set(txt2)

if txt_s1 == txt_s2:
  print(f"{txt1} and {txt2} are anagrams ")
else:
  print(f"{txt1} and {txt2} are not anagrams ")
```

### Question 6: Write a Python program to remove all the vowels from a string.

- `sample input : `
    - `var = 'python'`
- `sample output :`
    - `result : pythn , because vowels are [AEIOUaeiou]`

```python
# Please write The solution Here for above question 6:

txt = "Sehwag"
txt = txt.lower()

vowels = "aeiou"
new_txt = ""

for char in txt:
  if char not in vowels:
    new_txt = new_txt + char

print(new_txt)
```

### Question 7: Write a Python program to find the longest word in a string.

- `sample input : `
    - `var = 'python is Easy Language'`
- `sample output :`
    - `result : Language`

```python
# Please write The solution Here for above question 7:

txt = "python is Easy Language"

words = txt.lower().split()  #['python', 'is', 'easy', 'language']
#print(words)
len_dict = {}

for w in words:
  len_dict[w] = len(w)

#print(len_dict)

max_key = max(len_dict, key=len_dict.get)
print(max_key)
```

### Question 8: Write a Python program to capitalize the first letter of each word in a string [dont use built-in-function called title].

- `sample input : `
    - `var = 'python is easy'`
- `sample output :`
    - `result : Python Is Easy`

```python
# Please write The solution Here for above question 8:

txt = "rAghAV plays Cricket and badMintion"
words = txt.split()  #['python', 'is', 'easy', 'language']
new_txt = ""

for w in words:
  w = w[0].upper() + w[1:]
  new_txt = new_txt + " " + w

print(new_txt)
```

### Question 10: write a python programme to find the sum of all the even characters based on Ascii values

- `sample input : `
    - `var = 'sharuk khan'`
- `sample output :`
    - `result : 464`


```python
# Please write The solution Here for above question 10:

txt = "sharuk khan"
sum  = 0

for char in txt:
  i = ord(char)
  if (i % 2 == 0):  #to check whether value of a character is even
    sum =  sum + i

print(sum)
```
