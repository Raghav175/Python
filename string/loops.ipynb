## **for loop**

**1.	Write a program using a for loop to print each character of a string on a new line.**

```python
name = "Raghav"

for char in name:
  print(char)
```

**3.	Write a for loop program to count how many vowels are in a given string.**

```python
s = input()

count = 0
vowels = "aeiou"

for char in s:
  if char.lower() in vowels:
    count = count + 1
print(count)
```

**15.	Write a for loop program to count how many alphabetic characters, digits, and special characters are in a string.**


```python
name = "Raghav Surisetti#2154"

alphabet_count = 0
digit_count = 0
spl_count = 0

name_lower = name.lower()

for char in name_lower:
  if char.isalpha() ==  True:
    alphabet_count += 1
  elif char.isdigit() == True:
    digit_count += 1
  else:
    spl_count += 1

print(alphabet_count)
print(digit_count)
print(spl_count)
```

## **while loop**

**2.	Write a program using a while loop to print each character of a string with its index number.**

```python
name = "Raghav"
i = 0
while i < len(name):
  print(i, name[i])
  i = i + 1
```

**4.	Write a while loop program to reverse a string manually.**

```python
s = "Raghav"
rev_s = ""

l = len(s) # 6
i = 0
while i < l: # 0 < 6 i.e., 5 iterations for while
  rev_s = rev_s + s[l-i-1]
  i = i + 1

print(rev_s)
```

**6.	Write a program using a while loop to count how many times a specific character appears in a string.**

```python
s = "Raghav"

count = 0

char = input()


l = len(s) # 6
i = 0
while i < l: # 0 < 6 i.e., 5 iterations for while
  if s[i] == char:
    count += 1
  i = i + 1

print(f"{char} is repeated {count} times")
```

**8.	Write a while loop program that stops printing characters when it encounters the letter a.**


```python
s = "Raghav"

l = len(s) # 6

i = 0

while i < l: # 0 < 6 i.e., 6 iterations for while
  print(s[i])
  if s[i] == 'a':
    break
  i = i + 1
```

**10.	Write a while loop program to print alternate characters of a string**


```python
s = "Raghav Surisetti"

l = len(s) # 16

i = 0

while i < l: # 0 < 16 i.e., 16 iterations for while
  print(s[i])
  i = i + 2
```

**12.	Write a program using a while loop to copy characters of one string into another new string.**


```python
s = "Raghav"

l = len(s)

new_s = ""

i = 0

while i < l: # 0 < 6 i.e., 5 iterations for while
  new_s += s[i]
  i = i + 1

print(new_s)

#shallow copy check
print(id(s))
print(id(new_s))
```

**14.	Write a while loop program that prints the first half of a string.**

```python
s = "Surisetti's"

l = len(s) # 16

sub_s = ""
i = 0
while i < int(l/2): # 0 < 9 i.e., 8 iterations for while
  sub_s += s[i]
  i = i + 1

print(sub_s)
```
