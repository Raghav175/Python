**1. Take a string from user and print its length.**

```python
s = "Raghav loves to watch BIGGBOSS"

print(len(s))
```

**2. Input a string and print first and last character.**

```python
s = "Raghav loves to watch BIGGBOSS"

print(s[0]) # forward indexing starts with 1 & prints 1st character
print(s[-1]) #backward indexing starts with -1 & prints last character
```

**3. Input a string and print character at index 2 and −2.**

```python
s = "Raghav loves to watch BIGGBOSS"

print(s[2])
print(s[-2])
```

**4. Input a string and print the middle character(s).**

```python
s = "Sehwag"

l = len(s)

if (l % 2 == 0):
  print(s[int(l/2)])
  print(s[int(l/2) - 1])
else:
  print(s[l//2])
```

**12. Check if a string starts with "A" or ends with "Z".**

```python
s = "Ajinkya Rahane"

starts_with_a = s.startswith('A')
print(starts_with_a)
```

```python
s = "DragonBall-Z"

ends_with_a = s.endswith('Z')
print(ends_with_a)
```

**13. Count occurrences of each character in string using count().**

```python
s = "Rahane"

for char in s:
  print(f"count of {char} is : {s.count(char)}")
```

**14. Find position of substring in given string.**

```python
s = "Rahane"

print(s.find('ha')) #returns lower index or index of 1st character in string

print(s.index('ha'))
```

**15. Split a string into words and print each word separately (no loop — use index access methods).**

```python
s = "Raghav loves to watch BIGGBOSS"

print(s.split())
```

```python
s = "Raghav loves to watch BIGGBOSS"
words = s.split()
print(*words, sep='\n') #gemini
```

**17. Swap first and last characters of string using slicing.**

```python
s = "Raghav"

t = s[0]  #R

s1 = s.replace(s[0], s[-1])
print(s1)

s2 = s1.replace(s[-1], t)
print(s2)
```

```python
s = "Virat Kohli"   #gemini

swapped_string = s[-1] + s[1:-1] + s[0]

print(swapped_string)
```
