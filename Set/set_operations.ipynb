```python
set_a = {5, 10, 30, 50, 60}

set_a.add(100)

print(set_a)
```

**5.	Check if an element exists in a set.**

**6.	Find union of two sets.**

```python
set_a = {5, 10, 30, 50, 60}
set_b = {8, 15, 50, 10}

set_c = set_a.union(set_b)

print(set_c)
```

**7.	Find intersection of two sets.**

```python
set_a = {5, 10, 30, 50, 60}
set_b = {8, 15, 50, 10}

set_c = set_a.intersection(set_b)

print(set_c)
```

**8.	Find difference between two sets.**

```python
set_a = {5, 10, 30, 50, 60}
set_b = {8, 15, 50, 10}

set_c = set_a.difference(set_b)

print(set_c)
```

**9.	Find symmetric difference between two sets.**

```python
set_a = {5, 10, 30, 50, 60}
set_b = {8, 15, 50, 10}

set_c = set_a.symmetric_difference(set_b)
# Alternatively, you can use the ^ operator:
# set_c = set_a ^ set_b

print(set_c)
```

**11.	Create a frozen set from a list.**

**13.	Check if one set is subset of another.**


```python
set_a = {5, 10}
set_b = {5, 10, 30, 50, 60}

if set_a.intersection(set_b) == set_a:
  print(f"{set_a} is subset of {set_b}")

print(set_a.issubset(set_b))
```

**14.	Check if one set is superset of another.**

```python
set_a = {5, 10}
set_b = {5, 10, 30, 50, 60}

print(set_b.issuperset(set_a))
```

**15.	Check if two sets are disjoint.**

```python
set_a = {11, 15}
set_b = {5, 10, 30, 50, 60}

print(set_a.intersection(set_b))
print("sets are disjoint")
```
