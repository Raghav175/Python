**10. Create a copy of a dictionary and verify both are independent.**

```python
dict_s = {"Raghavendra" : 100,
          "Manobhi" : 89,
          "Ishan" : 65
          }

dict_s_copy = dict_s.copy() #creates a shallow copy

print(dict_s_copy)

dict_sa = {
    "Manasa" : 95
}

dict_s_copy.update(dict_sa) #using update() on the dict_s_copy
print(dict_s_copy)

print(dict_s)
```

### As updating the copy doesn't have any affect on the original dict(dict_s), both are independent

**11. Use setdefault() to insert a key only if it doesnâ€™t exist.**


```python
dict_s = {
    "Raghavendra" : 100,
    "Manobhi" : 89,
    "Ishan" : 65
          }

dict_s.setdefault("Manasa", 95) #inserting a new key into dict_s using setdefault
dict_s.setdefault("Raghavendra", 110)

print(dict_s)
```

**19. Use dictionary comprehension (without loops) to create a dictionary from a string using dict(zip()).**


```python
s = "raghav"


string_dict = {k : v for k,v in zip(s, range(len(s)))}

print(string_dict)
```
