**6. Create a shallow copy of a list.**

```python
lst_a = [3, 5, 8, 10, 15, "six"]

lst_b = lst_a.copy()
print(lst_b)

lst_b.append("Four") # appends "Four" to lsb_b
print(lst_b)
print(lst_a) # appending "Four to lst_b doesn't affect lst_a, so it's a shallow copy"
```

**Q)write a python program to reverse the order of the words in a given string**

```python
a = 'i.love.programming.language'
print(a)

f = a.split('.')
print(f)

t = []
l = len(f)

for i in range(l):
    t.append(f[l - (i+1)])

print('.'.join(t))
```
