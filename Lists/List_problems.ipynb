### Question 1: Write a Python program to count the number of elements in a list without using len.

### Question 4: Write a Python program to find the average of all the elements in a list.


- `sample input : `
    - `var = [1,2,3,4,2,5]`
- `sample output :`

```python
# Please write The solution Here for above question 4:
sum = 0
lst = [1,2,3,4,2,6]
l = len(lst)

for num in lst:
  sum += num
print("Average of numbers in the list", sum/l)
```

### Question 6:  Write a Python program to find the index of a specified element in a list..

- `sample input : `
    - `var = [1,2,3,4,2,5]`
- `sample output :`
    - `result : 3 index is 2`

```python
# Please write The solution Here for above question 6:

list_a = [1,2,3,4,2,5]
le = len(list_a)
n = int(input())

for i in range(le):
  if (list_a[i] == n):
    print(f"index of {n} is {i}")
```

### Question 7: Write a Python program to find which number is palindrome in below given list .

- `sample input : `
    - `var = [134,121,155]`
- `sample output :`
    - `result : 134 is not a palindome , 121 is a palindrome `

```python
# Please write The solution Here for above question 7:

lst =  [134,121,155]

for num in lst:
  if str(num) == str(num)[::-1]:
    print(f"{num} is palindrome")
  else:
    print(f"{num} is not a palindrome")
```

### Question 8: Write a Python program to find the common elements between two lists.

- `sample input : `
    - `var1 = [1,2,3,4,10]`
    - `var2 = [10,11,12,14]`
- `sample output :`
    - `result : 10`

```python
# Please write The solution Here for above question 8:

list_a = [1,2,3,4,10]
list_b = [10,11,12,14]

for num1 in list_a:
  for num2 in list_b:
    if (num1 == num2):
      print(f"The common element is: {num1}")
```

### Question 9: Write a Python program to remove the duplicates from a list

- `sample input : `
    - `var = [10,11,12,10,12,24]`
- `sample output :`
    - `result : [10,11,12,24]`

```python
# Please write The solution Here for above question 9:

list_z = [10,11,12,10,12,24]
list_a = []
list_b = []
for num in list_z:
  if (list_z.count(num) == 1):
    list_a.append(num)
  elif (list_z.count(num) > 1):
    list_b.append(num)

print(list_a + list(set(list_b)))
```
